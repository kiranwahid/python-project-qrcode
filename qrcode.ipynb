```python
# üì¶ For generating QR codes
import qrcode
# üì∑ For decoding QR codes using computer vision
import cv2
import numpy as np
# üñºÔ∏è For handling images
from PIL import Image
 # üñ•Ô∏è To display images in Jupyter
from IPython.display import display

# ‚úÖ Function to generate a QR code and save it as an image
def generate_qr_code(data, filename = "qrcode.png"):
    qr = qrcode.QRCode(
        # QR version (size). 1 is 21x21 matrix
        version = 1 ,
         # Size of each box in pixels
        box_size= 10,
        # Thickness of the border (quiet zone)
        border= 5)
    # ‚ûï Add your data to QR
    qr.add_data(data)
      # üîÑ Auto fit size based on data
    qr.make(fit=True)
     # üé® Colors
    img = qr.make_image(fill_color = "green" , back_color = "white")
     # üíæ Save image file
    img.save(filename)
     # üëÄ Display QR code (works in Jupyter)
    display(img)
    print(f"QR code saved as {filename}")

# ‚úÖ Function to decode a QR code image
def decode_qr_code(filename):
    # üì∑ Read the image using OpenCV
    img = cv2.imread(filename)
     # üîç Initialize QR code detector
    detector = cv2.QRCodeDetector()

    # üîç Decode the QR code (returns data, points, and straight_qrcode)
    data, _, _ = detector.detectAndDecode(img)

    # ‚úÖ Check and display the decoded data
    if data:

      print(f"‚úÖ Decoded data: {data} üéâ")
    else:
      print("‚ùå QR code could not be decoded.  üö´")

    # üß™ Example usage:
generate_qr_code("Successfully Generatedüåç", "my_qr.png")
decode_qr_code("my_qr.png")








```
